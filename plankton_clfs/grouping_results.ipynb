```python
predicted_labels = np.load("./ndsb_dataset_nounk/predicted_labels_079_valid.npy")
predictions= np.load("./ndsb_dataset_nounk/complete_predictions_079_valid.npy")

real_labels = np.load("./ndsb_dataset_nounk/real_labels_nounk.npy")

focus_measures = np.load("focus_valid_ndsb_nounk.npy")
```

```python
grouped_labels_real = group(real_labels)
grouped_labels_predicted = group(predicted_labels)
```

```python
np.save("./ndsb_dataset_nounk/labels_real_079_valid_grouped.npy", grouped_labels_real)
np.save("./ndsb_dataset_nounk/labels_predicted_079_valid_grouped.npy", grouped_labels_predicted)
```

```python
data = pd.DataFrame(predictions)
data['max'] = data.apply(max, axis=1)
data['real'] = grouped_labels_real
data['predicted'] = grouped_labels_predicted
data['lapv'] = focus_measures[:,0]
data['lapm'] = focus_measures[:,1]
data['teng'] = focus_measures[:,2]
```
